```python
import sys

sys.path.append('..')
from text2vec import SentenceModel, EncoderType
from text2vec import Word2Vec
import pickle
```

```python
def compute_emb(model):
    # Embed a list of sentences
    itemId_to_items = {}
    with open("data/itemid2item.pkl","rb") as f:
        itemId_to_items = pickle.load(f)
    item_ids = []
    titles = []
    for k, v in itemId_to_items.items():
        item_ids.append(k)
        titles.append(v[3])
    print(len(item_ids))
    print(len(titles))
    
    sentence_embeddings = model.encode(titles, show_progress_bar=True)
    print(type(sentence_embeddings), sentence_embeddings.shape)
    id_title_emb = []
    

    # The result is a list of sentence embeddings as numpy arrays
    for item_id, title, embedding in zip(item_ids, titles, sentence_embeddings):
        list = [item_id, title, embedding]
        id_title_emb.append(list)
    print(id_title_emb[:3])
    with open("emb/id_title_emb.pkl","wb") as f:
        pickle.dump(id_title_emb,f)
    
```

```python
t2v_model = SentenceModel("shibing624/text2vec-base-chinese",
                              encoder_type=EncoderType.FIRST_LAST_AVG)
compute_emb(t2v_model)
```
